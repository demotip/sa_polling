# 1 Generate Bursts Data Object

```python
date1 = '2014-03-01'
date2 = '2014-08-30'
start = datetime.strptime(date1, '%Y-%m-%d')
end = datetime.strptime(date2, '%Y-%m-%d')
step = timedelta(days=1)
bursts = {}
while start <= end:
    bursts[start.date()] = {'total':0.0, 'standard': 0.0, 'lottery': 0.0, 'subsidy': 0.0}
    start += step
```

```python
x = 0
conversations_df = pd.read_pickle('./LARGE_{}.pkl'.format(x))
conversations_df['timestamp'] = pd.to_datetime(conversations_df['timestamp'])
conversations_df['date'] = pd.DatetimeIndex(conversations_df['timestamp']).normalize()
date = datetime(2014,4,16)
def get_results(df):
    return df[
        (df['date'] == date) & 
        (df['poi'] == conversations_df['from_addr']) &
        (df['content'].notnull()) &
        (df['content'] != 'Session timeout.') &
        (df['content'] != 'User timeout') &
        (df['content'] != 'User timeout.') 
    ]
results = get_results(conversations_df)
```

```python
date1 = '2014-03-01'
date2 = '2014-08-30'
start = datetime.strptime(date1, '%Y-%m-%d')
end = datetime.strptime(date2, '%Y-%m-%d')
step = timedelta(days=1)
bursts = {}
while start <= end:
    bursts[start.date()] = {'total':0.0, 'control': 0.0, 'lottery': 0.0, 'subsidy': 0.0}
    start += step
    
for x in range(0,27):
    conversations_df = pd.read_pickle('./LARGE_{}.pkl'.format(x))
    conversations_df['timestamp'] = pd.to_datetime(conversations_df['timestamp'])
    conversations_df['date'] = pd.DatetimeIndex(conversations_df['timestamp']).normalize()
    for date, value in bursts.items():
        results = conversations_df[
            (conversations_df['date'] == date) & 
            (conversations_df['poi'] == conversations_df['from_addr']) &
            (conversations_df['content'].notnull()) &
            (conversations_df['content'] != 'Session timeout.') &
            (conversations_df['content'] != 'User timeout') &
            (conversations_df['content'] != 'User timeout.') 
        ]
        bursts[date]['total'] += len(results)
        bursts[date]['standard'] += len(results[results['to_addr'] == '*120*7692*2#'])
        bursts[date]['lottery'] += len(results[results['to_addr'] == '*120*7692*3#'])
        bursts[date]['free'] += len(results[results['to_addr'] == '*120*4729#'])
    print(x)
```
