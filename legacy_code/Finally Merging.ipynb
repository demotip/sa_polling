```python
def merge(merge_list):
    rs = []
    for row in merge_list.iterrows():
        conn, db, conversations, raw = mongo_connect()
        mxit_number = row[1]['mxit']
        phone = row[1]['phone_number']
        print('merging: {} {}'.format(mxit_number, phone))
        
        # save old
        try:
            c = conversations.find({'poi': mxit_number})
        finally:
            conn.close()
        
        old_mxit_convo = pd.DataFrame(list(c))
        old_mxit_convo.to_csv('./old/csv/old_mxit_convo_{}.csv'.format(mxit_number))
        old_mxit_convo.to_pickle('./old/pickles/old_mxit_convo_{}.pkl'.format(mxit_number))
        
        # update to new
        try:
            results = conversations.update_many({'poi': mxit_number}, {'$set': {'poi': phone}})
        finally:
            conn.close()
        print(results.raw_result)
        results_df = pd.DataFrame(results.raw_result,index=[mxit_number])
        results_df.to_pickle('./old/results/mxit_result_{}_{}.pkl'.format(mxit_number,phone))
        
        
        rs.append(results_df)
    return rs
```
