```python
df2.index = pd.to_datetime(df2.index)
range_1 = 35
range_2 = -112
# plt.plot(df2.index[range_1:range_2], df2['total'].iloc[range_1:range_2])
fig, ax = plt.subplots()
ax.plot(df2.index[range_1:range_2], df2['control'].iloc[range_1:range_2], label='lottery')
ax.plot(df2.index[range_1:range_2], df2['lottery'].iloc[range_1:range_2], label='control')
ax.plot(df2.index[range_1:range_2], df2['subsidy'].iloc[range_1:range_2], label='subsidy')

ax.set_ylabel('# responses')
ax.set_title('Number of Reponses')
# xticks = df2.index[range_1:range_2:2]
plt.xticks(rotation=45, fontsize=8)
ax.tick_params(axis='y',which='minor',bottom='on')
plt.yticks(pd.np.arange(0, 9e4, 1e4), ('0', '10k', '20k', '30k', '40k', '50k', '60k', '70k', '80k') )
plt.legend(loc=1)
plt.grid()
plt.savefig('./my_fig.png', dpi=100)
```
